### Ejercicio 2

Haga una funcion que obtenga la derivada numerica de una funcion utilizando f'(x) = f(x+h) - f(x) / h y utilice h para obtener buena precision. haga la prueba de que su funcion hace bien las cosas haciendo la grafica de f(x) = x^3 - 5x + 2, su derivada analitica y numerica. (las 3 graficas en la misma figura). (3 ptos)

```python
def prueba(x): 
    return (x*x*x - 5*x + 2)

def derivada_analitica_prueba(x):
    return 3*x*x - 5

def derivada_numerica(f, x, h=0.0001):
    return (f(x+h)-f(x))/h

#Declaramos la funcion de prueba sugerida, ponemos su derivada analitica que calculamos directamente
#y ponemos la funcion de derivada numerica
```

```python
fig=figure(figsize=(15,10))
ax = fig.add_subplot(111)
grid(True)
ax.set_title('Derivacion Numerica y Analitica',fontsize=20),xlabel('$x$',fontsize=20),ylabel('$y$',fontsize=20)

#Graficamos los resultados con su leyenda

x = linspace(0,10)
ax.plot (x, prueba(x), label="Funcion Prueba")
ax.plot (x, derivada_analitica_prueba(x), label="Derivada Analitica")
ax.plot (x, derivada_numerica(prueba, x), '+', label="Derivada Numerica")
ylim([-10,400]),grid(True)
ax.legend()
show()
```

### Ejercicio 5

Haga un programa que calcule la inversa de una matriz de 2x2.  Nota:  Si la matriz no tiene inversa, la funcion debera imprimir la leyenda \No tiene matriz inversa". (2 ptos)

```python
#Declaramos la funcion para invertir matrices, tomandolos como arrays de 2 dimensiones
#m = [[a,b],[c,d]]

def matriz_inversa(m):
    det = (m[0][0]*m[1][1]) - (m[0][1]*m[1][0])
    if det != 0:
        inversa = [[(1./det)*m[1][1], (-1./det)*m[0][1]], [(-1./det)*m[1][0], (1./det)*m[0][0]]]
    else:
        inversa = "No tiene matriz inversa"
    return inversa
```

```python
ejemplo=[[1,2],[3,4]]
matriz_inversa(ejemplo)
```

```python
ejemplo2=[[0,0],[0,0]]
matriz_inversa(ejemplo2)
```

### Ejercicio 6

Haga una funcion que se llame suma_numeros y que realice la suma de todos los digitos de un numero introducido.  

Ejemplo:sumanumeros(39474)
el resultado es: 27. (3 ptos)

```python
def suma_numeros(n):
    #Iteramos, cada vez tomando nuestro numero entero modulo 10 (le quitamos el decimal menos significativo), lo agregamos
    #a nuestro resultados, y hacemos division de piso (sacando el digito menos significativo)
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    return s
```

```python
suma_numeros(39474)
```

```python
#Euler
def euler(funcion,p_ini,t_final,h=0.01):
    tiempos = arange(p_ini[0], t_final+h,h)
    #y = zeros(len(tiempos))
    y = [0]*len(tiempos)
    y[0] = p_ini[1]
    for i in range (len(tiempos)-1):
        y[i+1] = y[i] + funcion(y[i],tiempos[i])*h
    solucion = [tiempos, y]
    return solucion
```

Esta es la ecuacion diferencial a resolver:
$f'(x) = f(x)$

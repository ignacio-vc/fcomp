```python
from JSAnimation import IPython_display
from matplotlib import animation
```

```python
def osc(y_v, t, omega2=1., mu=0):
    #y_v[0] es y, y_v[1] es v
    dy = y_v[1]
    dv = -omega2*sin(y_v[0]) - mu*y_v[1]
    return array([dy,dv])
```

```python
t = linspace(0,100,500)
inicio = [0.,1.]
sol = odeint(osc, inicio, t, args=(1,0.1))
xlim([-1,1]),ylim([-1,1])
plot(sol[:,0], sol[:,1])
```

```python
y,v = sol[:,0],sol[:,1]
#Pendulo. Asumiendo long cuerda l=1
pos_x, pos_y = sin(y), -cos(y)

figura = figure()
ejes = axes(xlim=(-1,1),ylim=(-1,1) )

punto, = ejes.plot([],[],'-o')
tache, = ejes.plot([0],[0],'x')

def el_movimiento(i):
    punto.set_data([0,pos_x[i]], [0,pos_y[i]])
    return punto,

animation.FuncAnimation(figura,el_movimiento,frames=len(y),interval=40)
```

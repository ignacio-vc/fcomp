```python
from pylab import *
from scipy.integrate import *
%matplotlib inline
```

```python
def ec_mov(x_p,t,epsilon=1.):
    # x_p = x,y,px,py
    r32 = (x_p[0]**2 + x_p[1]**2 )**(1.5)
    dx = x_p[2]
    dy = epsilon*x_p[3] 
    dpx = -x_p[0]/r32
    dpy = -x_p[1]/r32
    return array([dx,dy,dpx,dpy])

def energia(x_p, epsilon=1.):
    # x_p = x,y,px,py
    r = (x_p[0]**2 + x_p[1]**2)**(0.5)
    E = 0.5*(x_p[2]**2 + epsilon*x_p[3]**2) - 1./r
    return E

def cond_ini(alfa,var=0.001):
    angulo = alfa+var*(2.*rand()-1.)
    x,y = cos(angulo),sin(angulo)
    return array([x,y,0.5,0.5])

#Momento en la direccion del radio
def p_rho(x_p):
    # x_p = x,y,px,py
    r = (x_p[:,0]**2 + x_p[:,1]**2)**(0.5)
    x,y,px,py = x_p[:,0],x_p[:,1],x_p[:,2],x_p[:,3]
    p_r = (x*px + y*py)/r
    return p_r,r
```

```python
t = linspace(0,10,300)
#energia debe de ser negativa
pto_ini = (0,1,.5,.5)
print "E= ", energia(pto_ini)
sol = odeint(ec_mov,pto_ini,t)
plot(sol[:,0],sol[:,1])

# sol = odeint(ec_mov, )
```

```python
def root_trayec(func,pto_i,ti,error=1e-5):
    while abs(pto_i[0])>error:
        tip1 = ti - (pto_i[0]/pto_i[2])
        paso = odeint(func,pto_i,[ti,tip1])
        ti,pto_i = tip1,paso[-1]
    return ti,pto_i
```
